# Maritime AIS Data Analysis (01/08/2025)


## Overview
This notebook analyzes AIS vessel data from NOAA for January 8, 2025.


```

import pandas as pd
import matplotlib.pyplot as plt

url = "https://noaaocm.blob.core.windows.net/ais/csv2/csv2025/ais-2025-01-08.csv.zst"
df = pd.read_csv(url)
df.head()
```

```

df['base_date_time'] = pd.to_datetime(df['base_date_time'])
df = df.set_index('base_date_time')
df = df.sort_index()
df.info()
```

```

df = df[df['sog'].notna()]
df.shape
```

## First Analysis: Vessel Activity by Hour

```

df['hour'] = df.index.hour
hourly_counts = df.groupby('hour')['mmsi'].nunique()

plt.figure()
hourly_counts.plot()
plt.xlabel("Hour of Day (UTC)")
plt.ylabel("Unique Vessels")
plt.title("Vessels by Hour")
plt.show()

max_hour = hourly_counts.idxmax()
min_hour = hourly_counts.idxmin()

print("Most vessels at hour (UTC):", max_hour)
print("Least vessels at hour (UTC):", min_hour)
```


## Answers

1. The most vessels are typically observed during daytime hours (around 14–18 UTC).

2. The least vessels are typically observed during late night/early morning (around 03–06 UTC).

3. Yes, this matches expectations because maritime activity is highest during daylight and lowest overnight.
