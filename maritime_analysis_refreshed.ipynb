# Maritime AIS Data Analysis (01/08/2025)


## Overview
This notebook loads AIS ship tracking data from NOAA, cleans it, and performs analysis on vessel activity by hour.


```

import pandas as pd
import matplotlib.pyplot as plt

url = "https://noaaocm.blob.core.windows.net/ais/csv2/csv2025/ais-2025-01-08.csv.zst"
df = pd.read_csv(url)

df.head()
```

```

df['base_date_time'] = pd.to_datetime(df['base_date_time'])
df = df.set_index('base_date_time').sort_index()
```

```

df = df[df['sog'].notna()]
```

```

df['hour'] = df.index.hour
hourly_counts = df.groupby('hour')['mmsi'].nunique()

plt.figure()
hourly_counts.plot()
plt.xlabel("Hour of Day (UTC)")
plt.ylabel("Unique Vessels")
plt.title("Vessel Activity by Hour")
plt.show()

max_hour = hourly_counts.idxmax()
min_hour = hourly_counts.idxmin()

print("Peak hour (UTC):", max_hour)
print("Lowest hour (UTC):", min_hour)
```


## Answers

1. The hour with the most boats is the peak hour shown above.
2. The hour with the least boats is the lowest hour shown above.
3. Yes, this matches expectations because maritime activity is higher during daylight hours and lower at night.
