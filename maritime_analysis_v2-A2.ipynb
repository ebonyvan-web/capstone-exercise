# AIS Maritime Analysis (01/08/2025)

This notebook loads AIS vessel data, cleans it, and performs analysis on vessel activity and a specific ship (Henry Hudson).

```
import pandas as pd

# Load dataset (compressed CSV)
url = "https://noaaocm.blob.core.windows.net/ais/csv2/csv2025/ais-2025-01-08.csv.zst"
df = pd.read_csv(url)

df.head()
```

```
# Convert base_date_time to datetime and set as index
df['BaseDateTime'] = pd.to_datetime(df['BaseDateTime'])
df = df.set_index('BaseDateTime').sort_index()

df.info()
```

```
# Clean data: remove rows where SOG is null
df = df[df['SOG'].notnull()]

df.shape
```

```
# First Analysis: Number of unique vessels by hour

df['hour'] = df.index.hour

vessels_per_hour = df.groupby('hour')['MMSI'].nunique()

vessels_per_hour
```

```
# Plot vessels per hour
import matplotlib.pyplot as plt

plt.figure()
vessels_per_hour.plot(kind='line')
plt.xlabel("Hour (UTC)")
plt.ylabel("Number of Unique Vessels")
plt.title("Vessel Activity by Hour (UTC)")
plt.show()

print("Peak hour:", vessels_per_hour.idxmax())
print("Least active hour:", vessels_per_hour.idxmin())
```

## Second Analysis: Henry Hudson (MMSI: 366651000)

```
# Filter for Henry Hudson
henry = df[df['MMSI'] == 366651000].copy()

# Ensure sorted index
henry = henry.sort_index()

# Acceleration (change in SOG)
henry['acceleration'] = henry['SOG'].diff()

# Duration (time difference)
henry['duration'] = henry.index.to_series().diff()

henry.head()
```

```
# Filter where acceleration > 0 (engine active)
active = henry[henry['acceleration'] > 0]

# Total active duration
total_active_time = active['duration'].sum()

total_active_time
```

### Conclusion
- Peak vessel activity hour: see printed output
- Least vessel activity hour: see printed output
- Total engine active time for Henry Hudson: shown above
