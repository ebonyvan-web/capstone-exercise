# AIS Maritime Analysis (01/08/2025)

```

import pandas as pd
import matplotlib.pyplot as plt

url = "https://noaaocm.blob.core.windows.net/ais/csv2/csv2025/ais-2025-01-08.csv.zst"
df = pd.read_csv(url)

df['base_date_time'] = pd.to_datetime(df['base_date_time'])
df = df.set_index('base_date_time').sort_index()

df = df[df['sog'].notna()]
```

```

# First Analysis
df['hour'] = df.index.hour
hourly = df.groupby('hour')['mmsi'].nunique()

plt.figure()
hourly.plot()
plt.xlabel("Hour (UTC)")
plt.ylabel("Unique Vessels")
plt.title("Vessels by Hour")
plt.show()

hourly.idxmax(), hourly.idxmin()
```


## Answers
1. Most boats: ~14–18 UTC  
2. Least boats: ~03–06 UTC  
3. Yes — aligns with expected daytime activity patterns
